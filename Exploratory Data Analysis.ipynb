```python
import pandas as pd
import os
```

Merge the files

```python
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv('allData.csv')
```

```python
all_data = pd.read_csv('allData.csv')
all_data.head()
```

```python
#Drop null values
all_data = all_data.dropna(axis=0, ignore_index=True)
all_data.head()
```

Augment data with additional columns

Add monnth column

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

Convert column to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.head()
```

Task 3: Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

Add a city column

```python
def get_city(address):
    return address.split(',')[1]

def get_country(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ', ' + get_country(x))
all_data.head()
```

Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month')['Sales'].sum()
results
```

December has the highest sales with $4613443.34.

```python
import matplotlib.pyplot as plt

months = range(1,13)
plt.bar(months, results)
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

Question 2: What city had the highest number of sales?

```python
city_sale = all_data.groupby('City').sum()
city_sale
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')]
plt.bar(cities, city_sale['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()
```

San Francisco has the highest sales with amount $8262203.91.

Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

Noon (12 PM) or 7 PM would be best time to advertise the product to maximize the likelihood of customer's buying product.

Question 4: What products are most often sold together?

```python
dup = all_data[all_data['Order ID'].duplicated(keep=False)]
dup['Grouped'] = dup.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
dup = dup[['Order ID', 'Grouped']].drop_duplicates()
dup.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in dup['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

Question 5: What product sold the most and why?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum(['Quantity Ordered'])

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered['Quantity Ordered'])
plt.xticks(products, rotation='vertical', size=8)
plt.xlabel('Product')
plt.ylabel('Number of Orders')
plt.show()
```

AAA Batteries (4-pack) is the most sold product among others.

```python
prices = all_data.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered['Quantity Ordered'], color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```

As the price increases, the quantity ordered decreases which could be a reason for AAA Batteries (4-pack) sold the most as its price is lowest.
